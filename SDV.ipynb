```python
import pandas as pd

real_data = pd.read_csv('Cancer_Data.csv')
real_data.head()
```

```python
from sdv.single_table import CTGANSynthesizer
```

```python
from sdv.metadata import SingleTableMetadata

Metadata = SingleTableMetadata()
Metadata.detect_from_dataframe(real_data)
```

```python
model = CTGANSynthesizer(metadata = Metadata)
model.fit(real_data)
```

```python
Synthetic_data = model.sample(num_rows = 569)
Synthetic_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Count the occurrences of each diagnosis category
diagnosis_counts = Synthetic_data['diagnosis'].value_counts()

# Create the bar chart
plt.figure(figsize=(8, 6))
bars = sns.barplot(x=diagnosis_counts.index, y=diagnosis_counts.values)

# Add labels and title
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.title('Distribution of Diagnosis Categories in Real Cancer Data')
plt.xticks(ticks=[0, 1], labels=['Benign (B)', 'Malignant (M)']) # Assuming 'B' and 'M' are the labels

for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), bar.get_height(),
             ha='center', va='bottom')

# Display the plot
plt.show()
```

```python
from sdv.evaluation.single_table import get_column_plot

fig = get_column_plot(
    real_data=real_data,
    synthetic_data=Synthetic_data,
    column_name='diagnosis',
    metadata=Metadata
)

fig.show()
```

```python
from sdv.evaluation.single_table import evaluate_quality

quality_report = evaluate_quality(
    real_data,
    Synthetic_data,
    metadata=Metadata
)
```

```python
import pandas as pd

# Check for missing values in each column
missing_values = real_data.isnull().sum()

# Display columns with missing values
print("Columns with missing values:")
print(missing_values[missing_values > 0])
```

```python
import numpy as np

# Select only numerical columns for outlier detection
numerical_cols = real_data.select_dtypes(include=np.number).columns.tolist()

# Remove 'id' and 'Unnamed: 32' as they are not relevant for outlier detection
numerical_cols.remove('id')
if 'Unnamed: 32' in numerical_cols:
    numerical_cols.remove('Unnamed: 32')


print("Potential Outliers (using IQR method):")

# Iterate over numerical columns and find outliers
for col in numerical_cols:
    Q1 = real_data[col].quantile(0.25)
    Q3 = real_data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = real_data[(real_data[col] < lower_bound) | (real_data[col] > upper_bound)]

    if not outliers.empty:
        print(f"\nOutliers in '{col}':")
        display(outliers[['id', col]])
```

```python
# Drop the 'Unnamed: 32' column as it contains only missing values
real_data_cleaned = real_data.drop('Unnamed: 32', axis=1)

# Display the first few rows of the cleaned data
display(real_data_cleaned.head())
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train = Synthetic_data.drop('diagnosis', axis=1)
y_train = Synthetic_data['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
# Drop the 'Unnamed: 32' column from X_train and X_test
X_train_cleaned = X_train.drop('Unnamed: 32', axis=1)
X_test_cleaned = X_test.drop('Unnamed: 32', axis=1)

# Logistic Regression
lr_model = LogisticRegression(max_iter=1000) # Increased max_iter for convergence
lr_model.fit(X_train_cleaned, y_train)
lr_preds = lr_model.predict(X_test_cleaned)

# Support Vector Machine
svm_model = SVC()
svm_model.fit(X_train_cleaned, y_train)
svm_preds = svm_model.predict(X_test_cleaned)

# Decision Tree
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train_cleaned, y_train)
dt_preds = dt_model.predict(X_test_cleaned)
```

```python
# Evaluate Logistic Regression
print("--- Logistic Regression Evaluation ---")
print("Accuracy:", accuracy_score(y_test, lr_preds))
print(classification_report(y_test, lr_preds))

# Evaluate Support Vector Machine
print("\n--- Support Vector Machine Evaluation ---")
print("Accuracy:", accuracy_score(y_test, svm_preds))
print(classification_report(y_test, svm_preds))

# Evaluate Decision Tree
print("\n--- Decision Tree Evaluation ---")
print("Accuracy:", accuracy_score(y_test, dt_preds))
print(classification_report(y_test, dt_preds))
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_cleaned, y_train)
rf_preds = rf_model.predict(X_test_cleaned)

# Gradient Boosting
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train_cleaned, y_train)
gb_preds = gb_model.predict(X_test_cleaned)
```

```python
# Evaluate Random Forest
print("--- Random Forest Evaluation ---")
print("Accuracy:", accuracy_score(y_test, rf_preds))
print(classification_report(y_test, rf_preds))

# Evaluate Gradient Boosting
print("\n--- Gradient Boosting Evaluation ---")
print("Accuracy:", accuracy_score(y_test, gb_preds))
print(classification_report(y_test, gb_preds))
```

```python
# Prepare the real data for evaluation
X_real = real_data_cleaned.drop('diagnosis', axis=1)
y_real = real_data_cleaned['diagnosis']

# Make predictions on the real data using the trained Random Forest model
rf_real_preds = rf_model.predict(X_real)

# Evaluate the Random Forest model on the real data
print("--- Random Forest Evaluation on Real Data ---")
print("Accuracy:", accuracy_score(y_real, rf_real_preds))
print(classification_report(y_real, rf_real_preds))
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Get the probability scores for the positive class ('M' for Malignant)
# Logistic Regression
lr_probs = lr_model.predict_proba(X_test_cleaned)[:, 1]

# Support Vector Machine (requires probability=True during model initialization)
# Since probability=True was not set, we'll skip SVM for now or re-train it if needed
# For now, let's assume we would get probabilities like this if probability=True was set:
# svm_probs = svm_model.predict_proba(X_test_cleaned)[:, 1]

# Decision Tree
dt_probs = dt_model.predict_proba(X_test_cleaned)[:, 1]

# Random Forest
rf_probs = rf_model.predict_proba(X_test_cleaned)[:, 1]

# Gradient Boosting
gb_probs = gb_model.predict_proba(X_test_cleaned)[:, 1]

# Map diagnosis labels to numerical values for ROC curve calculation
y_test_numeric = y_test.apply(lambda x: 1 if x == 'M' else 0)

# Calculate ROC curve and AUC for each model
lr_fpr, lr_tpr, lr_thresholds = roc_curve(y_test_numeric, lr_probs)
lr_auc = auc(lr_fpr, lr_tpr)

dt_fpr, dt_tpr, dt_thresholds = roc_curve(y_test_numeric, dt_probs)
dt_auc = auc(dt_fpr, dt_tpr)

rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test_numeric, rf_probs)
rf_auc = auc(rf_fpr, rf_tpr)

gb_fpr, gb_tpr, gb_thresholds = roc_curve(y_test_numeric, gb_probs)
gb_auc = auc(gb_fpr, gb_tpr)


# Plot the ROC curves
plt.figure(figsize=(10, 8))
plt.plot(lr_fpr, lr_tpr, label=f'Logistic Regression (AUC = {lr_auc:.2f})')
# plt.plot(svm_fpr, svm_tpr, label=f'SVM (AUC = {svm_auc:.2f})') # Uncomment if SVM probabilities are available
plt.plot(dt_fpr, dt_tpr, label=f'Decision Tree (AUC = {dt_auc:.2f})')
plt.plot(rf_fpr, rf_tpr, label=f'Random Forest (AUC = {rf_auc:.2f})')
plt.plot(gb_fpr, gb_tpr, label=f'Gradient Boosting (AUC = {gb_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Models Trained on Synthetic Data')
plt.legend()
plt.grid(True)
plt.show()
```
